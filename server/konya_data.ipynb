```python
import pandas as pd
import numpy as np
```

```python
df_sulak_alan = pd.read_csv('../data/sulak_alan.csv', delimiter=';')
df_sulak_alan.dropna(inplace=True)
df_sulak_alan
```

```python
df_en_cok_yetistirilen = pd.read_csv('../data/en_cok_yetistirilen.csv', delimiter=';')
df_en_cok_yetistirilen.dropna(inplace=True)
df_en_cok_yetistirilen
```

```python
df_sulama_kaynagi_tipi = pd.read_csv('../data/sulama_kaynagi_tipi.csv', delimiter=';')
df_sulama_kaynagi_tipi.dropna(inplace=True)
df_sulama_kaynagi_tipi
```

```python
df_sulanabilir_alanda_yetistirilen = pd.read_csv('../data/sulanabilir_alanda_yetistirilen.csv', delimiter=';')
df_sulanabilir_alanda_yetistirilen.dropna(inplace=True)
df_sulanabilir_alanda_yetistirilen
```

```python
df_tarimsal_egitim = pd.read_csv('../data/tarimsal_egitim.csv', delimiter=';')
df_tarimsal_egitim.dropna(inplace=True)
df_tarimsal_egitim
```

```python
df_tarimsal_sulama_durum_bilgisi = pd.read_csv('../data/tarimsal_sulama_durum_bilgisi.csv', delimiter=';')
df_tarimsal_sulama_durum_bilgisi.dropna(inplace=True)
```

```python
df_tarimsal_sulama_kaynagi = pd.read_csv('../data/tarimsal_sulama_kaynagi.csv', delimiter=';')
df_tarimsal_sulama_kaynagi.dropna(inplace=True)
df_tarimsal_sulama_kaynagi
```

```python

merged_df = pd.merge(df_tarimsal_sulama_durum_bilgisi, df_sulama_kaynagi_tipi, on=['ilce_adi', 'mahalle_adi'])
merged_df = pd.merge(merged_df, df_en_cok_yetistirilen, on=['ilce_adi', 'mahalle_adi'])
merged_df = pd.merge(merged_df, df_sulanabilir_alanda_yetistirilen, on=['ilce_adi', 'mahalle_adi'])
merged_df = pd.merge(merged_df, df_tarimsal_egitim, on=['ilce_adi', 'mahalle_adi'])
merged_df = pd.merge(merged_df, df_sulak_alan, on=['ilce_adi', 'mahalle_adi'])
merged_df = pd.merge(merged_df, df_tarimsal_sulama_kaynagi, on=['ilce_adi', 'mahalle_adi'])

merged_df
```

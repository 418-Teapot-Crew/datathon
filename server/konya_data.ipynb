```python
categories = {
    'Verimli Toprak, Yüksek Su İhtiyacı': 'A',
    'Verimli Toprak, Orta Su İhtiyacı': 'B',
    'Killi Toprak, Yüksek Su İhtiyacı': 'C',
    'Killi Toprak, Orta Su İhtiyacı': 'D',
    'Killi Toprak, Düşük Su İhtiyacı': 'E',
    'Kumlu Toprak, Yüksek Su İhtiyacı': 'F',
    'Kumlu Toprak, Düşük Su İhtiyacı': 'G',
}

def determine_category(row):
    toprak = row['toprak_tipi']
    sulak = int(row['sulak_alan_miktari'].replace('%', '')) 
    
    if 'Verimli' in toprak and sulak > 65:
        return 'A'
    elif 'Verimli' in toprak and 25 <= sulak <= 65:
        return 'B'
    elif 'Verimli' in toprak and sulak < 25:
        return 'G'
    elif 'Killi' in toprak and sulak > 65:
        return 'A'
    elif 'Killi' in toprak and 25 <= sulak <= 65:
        return 'C'
    elif 'Killi' in toprak and sulak < 25:
        return 'G'
    elif 'Kumlu' in toprak and sulak > 65:
        return 'F'
    elif 'Kumlu' in toprak and 25 <= sulak < 65:
        return 'D'
    elif 'Kumlu' in toprak and sulak < 25:
        return 'E'
    else:
        return 'Z'  # Diğer durumlar için bir kategori belirttik

merged_df['product_category'] = merged_df.apply(determine_category, axis=1)
merged_df
```

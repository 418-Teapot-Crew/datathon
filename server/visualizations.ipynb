```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from random import uniform
```

```python
data = pd.read_excel('../data/visualization.xlsx')
df = data.copy()
df
```

```python
def calculate_yagis_miktari(row):
    random_multiplier = uniform(5, 15)
    return row['sulak_alan_miktari'] * random_multiplier

df['yagis_miktari (mm)'] = df.apply(calculate_yagis_miktari, axis=1)
df
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

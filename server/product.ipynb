```python
import pandas as pd
import numpy as np
```

```python
df_sulanabilir_alanda_yetistirilen = pd.read_csv('../data/sulanabilir_alanda_yetistirilen.csv', delimiter=';')
df_sulanabilir_alanda_yetistirilen.dropna(inplace=True)
df_sulanabilir_alanda_yetistirilen
```

```python
unique_products = []
for products in df_sulanabilir_alanda_yetistirilen['sulanabilir_alanda_yetistirilen_urunler'].unique():
    ayrilmis_liste = products.split(',')
    for product in ayrilmis_liste:
        trimmed_product = product.strip()  
        if trimmed_product not in unique_products and trimmed_product != 'YOK': 
            unique_products.append(trimmed_product)

print(unique_products)
df_product = pd.DataFrame(columns=['product'])
df_product['product'] = unique_products
df_product
```

```python
categories_desc = {
    'Verimli Toprak, Yüksek Su İhtiyacı': ['SİYAH HAVUÇ', 'HAVUÇ', 'MISIR', 'ŞEKER PANCARI', 'KAVUN'],
    'Verimli Toprak, Orta Su İhtiyacı': ['PATATES', 'MARUL', 'DOMATES', 'BİBER', 'SALATALIK'],
    'Killi Toprak, Yüksek Su İhtiyacı': ['KARPUZ', 'YONCA', 'LAHANA', 'AYÇİÇEĞİ'],
    'Killi Toprak, Orta Su İhtiyacı':  ['ASPİR', 'ÇEREZLİK KABAK', 'ARPA', 'BUĞDAY', 'ÇİLEK'],
    'Killi Toprak, Düşük Su İhtiyacı': ['HAŞHAŞ', 'REZENE', 'KANOLA', 'KİMYON'],
    'Kumlu Toprak, Yüksek Su İhtiyacı': ['MERCİMEK', 'KABAK', 'NOHUT', 'FASULYE', 'PIRASA'],
    'Kumlu Toprak, Düşük Su İhtiyacı': ['ŞARIMSAK', 'SOĞAN', 'CEVİZ'],
}


categories = {
    'Verimli Toprak, Yüksek Su İhtiyacı': 'A',
    'Verimli Toprak, Orta Su İhtiyacı': 'B',
    'Killi Toprak, Yüksek Su İhtiyacı': 'C',
    'Killi Toprak, Orta Su İhtiyacı': 'D',
    'Killi Toprak, Düşük Su İhtiyacı': 'E',
    'Kumlu Toprak, Yüksek Su İhtiyacı': 'F',
    'Kumlu Toprak, Düşük Su İhtiyacı': 'G',
}
```

```python
combined_data = []

for category_desc, products in categories_desc.items():
    category_code = categories[category_desc]
    combined_data.extend([(product, category_desc, category_code) for product in products])

df_product_category = pd.DataFrame(combined_data, columns=['Product', 'Category_Desc', 'Category'])
df_product_category
```

```python
df_product_category.to_csv('../data/product_category.csv', index=False)
```

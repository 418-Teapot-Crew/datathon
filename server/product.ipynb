```python
import pandas as pd
import numpy as np
```

```python
df_sulanabilir_alanda_yetistirilen = pd.read_csv('../data/sulanabilir_alanda_yetistirilen.csv', delimiter=';')
df_sulanabilir_alanda_yetistirilen.dropna(inplace=True)
df_sulanabilir_alanda_yetistirilen
```

```python
unique_products = []
for products in df_sulanabilir_alanda_yetistirilen['sulanabilir_alanda_yetistirilen_urunler'].unique():
    ayrilmis_liste = products.split(',')
    for product in ayrilmis_liste:
        trimmed_product = product.strip()  
        if trimmed_product not in unique_products and trimmed_product != 'YOK': 
            unique_products.append(trimmed_product)

print(unique_products)
df_product = pd.DataFrame(columns=['product'])
df_product['product'] = unique_products
df_product
```

```python
combined_data = []

for category_desc, products in categories_desc.items():
    category_code = categories[category_desc]
    combined_data.extend([(product, category_desc, category_code) for product in products])

df_product_category = pd.DataFrame(combined_data, columns=['Product', 'Category_Desc', 'Category'])
df_product_category
```
